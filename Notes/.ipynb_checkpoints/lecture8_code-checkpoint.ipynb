## Last time:
Monte Carlo frequentist approach
Model: $y = 1 + 2 x$
$x: 1 - 4$, $\sigma$
$y = a + b x$, $\chi^2 = \displaystyle \sum_{i = 1}^N \frac{\left(y_i - a - b x_i\right)^2}{\sigma_i^2}$
$a$, $b$ were found to be highly anti-correlated

The problem with this method is that if you had millions of points with non-Gaussian errors, this becomes extremely difficult. Want to find different approaches as number of points grows.

We histogrammed $\chi^2$/dof, where dof = N - M, with M = 2, and we should find that
$$P(\chi^2) = \frac{(\chi^2)^{(k/2) - 1} \mathrm{e}^{-(\chi^2)/2}}{2^{k/2} \Gamma\left(\frac{k}{2}\right)}$$
mean = k, variance = 2k

Suppose we have 100 degrees of freedom, and we expect k = 1, get $\chi^2$/dof = 3, no way to know if this is good or not because we only have one realization (license plate paradox). You need to repeat the experiment many times to get a distribution. If the distribution is shifted $\rightarrow$ bad fit, if the distribution matches what you expect $\rightarrow$ good fit.

Example: Plots of $\chi^2$/dof for 3 neutron stars (*real data*). Vast majority follow the expected distribution, for most of the data the model does a very good job. This is what you want to do when using a $\chi^2$ test for data!

Two ways to get a bad $\chi^2$: mess up model description, or mess up errors. Can not tell which of the two happened just by looking at the $\chi^2$ distribution. There are signs to tell if it is one or the other:

* A bad fit (wrong model), the deviation of the data from the model will be systematic. Plotting the residuals (difference between data and model divided by the error) can tell you if this is the case. Correct model should have random distribution of residual points.
* Correct errors should produce residuals near 1.

Try using 100 data points, using model $y = 1 + 2 x + 0.2 x^2$, but try to fit with $y = a + b x$, $\sigma = 0.05$, $x$ drawn from 1-4. Plot residual for one of the realizations. For many realizations, plot the distribution of the $\chi^2$/dof and the values of $a$ and $b$ to see if they are around 1 and 2, respectively.

```python
Ndata = 100
Nreal = int(1e5)
np.random.seed(0)
y = np.empty((Nreal, Ndata))
afound, bfound, chisq = np.empty(Nreal), np.empty(Nreal), np.empty(Nreal)
# Get one realization of the data
x = np.linspace(1, 4, num=Ndata)
sigma_y = 0.05
areal, breal, creal = 1, 2, 0.2
z = utilities.get_fake_gaussian_data(Ndata)[0]
y[0] = areal + breal*x + creal*np.power(x, 2) + sigma_y*z.copy()
del z
# Find the best fit parameters for this realization
Sx = np.sum(x/np.power(sigma_y,2))
Sy = np.sum(y[0]/np.power(sigma_y,2))
S = Ndata/np.power(sigma_y,2)
t = (x - (Sx/S))/sigma_y
Stt = np.sum(np.power(t,2))
bfound[0] = np.sum(t*y[0]/sigma_y)/Stt
afound[0] = (Sy - Sx*bfound[0])/S
# Find the value of chi^2
chisq[0] = np.sum(np.power(y[0] - afound[0] - bfound[0]*x, 2)/np.power(sigma_y, 2))
# Make the plot with the residuals
fig1 = plt.figure(figsize=(10,6), facecolor='w')
frame1=fig1.add_axes((.1,.3,.8,.6))
frame1.set_ylabel('y')
plt.errorbar(x, y[0], yerr=sigma_y, fmt='ro')
plt.plot(x, afound[0] + bfound[0]*x, 'b-')
frame1.set_xticklabels([])
plt.grid()
residual = (y[0] - afound[0] - bfound[0]*x)/sigma_y
frame2=fig1.add_axes((.1,.1,.8,.2))
frame2.set_xlabel('x')
frame2.set_ylabel('residual')
plt.errorbar(x, residual, yerr=sigma_y, fmt='ro')
plt.grid()
print afound[0], bfound[0], chisq[0]
```

```python
# Do the other realizations
for i in range(1,Nreal) :
    z = utilities.get_fake_gaussian_data(Ndata)[0]
    y[i] = areal + breal*x + creal*np.power(x, 2) + sigma_y*z.copy()
    del z
    # Find the best fit parameters for this realization. Note Sx, S, t, and Stt don't change
    Sy = np.sum(y[i]/np.power(sigma_y,2))
    bfound[i] = np.sum(t*y[i]/sigma_y)/Stt
    afound[i] = (Sy - Sx*bfound[i])/S
    # Find the value of chi^2
    chisq[i] = np.sum(np.power(y[i] - afound[i] - bfound[i]*x, 2)/np.power(sigma_y, 2))
# Plot the histogram of chi^2/dof
dof = Ndata - 2
def P(x, k) :
    return ((np.power(x, (k/2.) - 1.)*np.exp(-x/2.))/(np.power(2, k/2.)*gamma(k/2.)))
plt.figure(figsize=(10,6), facecolor='w')
plt.xlabel('$\chi^2$/dof')
plt.hist(chisq/dof, bins=100, normed=True)
plt.plot(np.linspace(0.1, 12, num=1000), P(np.linspace(0.1, 12, num=1000), 1), 'r-')
plt.xlim([0, 12])
```

```python
# Plot a and b
samples = np.array([afound, bfound]).T
figure = corner.corner(samples, labels=['a', 'b'], truths=[areal, breal], show_titles=True, title_args={"fontsize": 16})
```

## Another way
Contours of constant posterior likelihood are contours of $\Delta \chi^2$. In a purely linear model, you can calculate those values of $\Delta \chi^2$ independent of the model. Just find minimum in (a, b) plane, then find $\Delta \chi^2(a, b) = \chi^2(a, b) - \chi^2_{\mathrm{min}}$.

P                         $\Delta \chi^2$ (1 dof)   $\Delta \chi^2$ (2 dof)
68.3% (1 $\sigma$)        1.0                       2.3
90%                       2.71                      4.61
95.4% (2 $\sigma$)        4.0                       6.17
99.7% (3 $\sigma$)        9.0                       11.8

Formally only true for linear models

Do only one realization of a true linear model, make a grid in (a, b) away from best fit values, then find $\Delta \chi^2$ at all values of (a, b) in grid.

```python
np.random.seed(0)
np.random.seed(1)
z = utilities.get_fake_gaussian_data(Ndata)[0]
ylin = areal + breal*x + sigma_y*z.copy()
Sx = np.sum(x/np.power(sigma_y,2))
Sy = np.sum(ylin/np.power(sigma_y,2))
S = Ndata/np.power(sigma_y,2)
t = (x - (Sx/S))/sigma_y
Stt = np.sum(np.power(t,2))
b0 = np.sum(t*ylin/sigma_y)/Stt
a0 = (Sy - Sx*b0)/S
chimin = np.sum(np.power(ylin - a0 - b0*x, 2)/np.power(sigma_y, 2))
agrid = np.linspace(0.94*a0, 1.06*a0, num=500)
bgrid = np.linspace(0.99*b0, 1.01*b0, num=500)
delchi = np.empty((agrid.size, bgrid.size))
for i, ai in zip(range(agrid.size), agrid) :
    for j, bj in zip(range(bgrid.size), bgrid) :
        delchi[i, j] = np.sum(np.power(ylin - ai - bj*x, 2)/np.power(sigma_y, 2)) - chimin
plt.figure(figsize=(10,6), facecolor='w')
CS = plt.contour(agrid, bgrid, delchi.T, levels=[2.3, 6.17, 11.8], origin='lower', colors=('r', 'g', 'b'))
plt.xlim([agrid.min(), agrid.max()])
plt.ylim([bgrid.min(), bgrid.max()])
plt.clabel(CS, inline=1)
plt.xlabel('a')
plt.ylabel('b')
plt.plot(a0, b0, 'ko')
```

## (Uncorrelated) errors in both directions

For point ($x_i, y_i$), have errors ($\sigma_{x_i}$, $\sigma_{y_i}$). Still want same properties for $\chi^2$ as before:

$$\chi^2 = \displaystyle \sum_{i = 1}^N \frac{\left(y_i - a - b x_i\right)^2}{weight}$$

Want a different weight. Var($y_i - a - b x_i$) = Var($y_i$) + $b^2$ Var($x_i$). So this variance is the weight, and 

$$\boxed{\chi^2 = \displaystyle \sum_{i = 1}^N \frac{\left(y_i - a - b x_i\right)^2}{\sigma_{y_i}^2 + b^2 \sigma_{x_i}^2}}$$

Everything else follows through except the equations for $\frac{\partial \chi^2}{\partial a} = 0$ and $\frac{\partial \chi^2}{\partial b} = 0$: no longer linear in $b$!

## Nonlinear models

Use a transformation of variables to convert to a linear model. E.g.
$$y = a b^x \Rightarrow y' = \log y = \log a + x \log b$$
$$y = a x^b \Rightarrow y' = \log y = \log a + b \log x = \log a + b x'$$
$$ y = a + b x^2 \mathrm{e}^{-x} = a + b x'$$

Fundamental error we will make: even if the uncertainties in $x$ and $y$ are Gaussian, the uncertainties in $x'$, $y'$ won't be Gaussian, and if you start combining $x$ and $y$ you get correlated errors. This violates the fundamental assumptions of $\chi^2$ statistics (Gaussian, uncorrelated errors)!

So you define a generalized $\chi^2$ given as (with errors only in one direction):
$$\chi^2 = \displaystyle \sum_{i = 1}^N \frac{\left[y_i - f\left(x_i; \vec{m}\right)\right]^2}{\sigma_i^2},$$

where $\vec{m}$ is the vector of parameters. We implicitly assume when finding the levels $\Delta \chi^2$ that we calculate the model and Taylor expand it around the best fit parameters $\vec{m_0}$:
$$f\left(x_i; \vec{m}\right) = f\left(x_i; \vec{m_0}\right) + \delta \vec{m} \nabla_m f$$

## Estimate errors

On average, (all errors the same)
$$\chi^2 = \frac{1}{\sigma^2} \displaystyle \sum_{i = 1}^N \left(y_i - a - b x_i\right)^2 = N_{\mathrm{dof}}$$

$$\boxed{\sigma^2 = \frac{1}{N_{\mathrm{dof}}} \displaystyle \sum_{i = 1}^N \left(y_i - a - b x_i\right)^2}$$

$$\chi^2 = \displaystyle \sum_{i = 1}^N \frac{\left(y_i - a - b x_i\right)^2}{\sigma_i^2 + \sigma_{sys}^2} = N_{\mathrm{dof}}$$

But this is only true on average, so must do this when we know the model agrees, don't try to inflate the errors to make it true!

## Unknown distribution of errors (Bootstrap)

Suppose we have N (large) data points. Make many realizations of N points (yes, the same as the number of points!) out of the data with replacement (repeating allowed) a large number of times. For each realization, minimize the $\chi^2$ to get the values of the parameters. The distribution of the parameters you get after all realizations is actually the true distribution. It requires a large number of realizations of a large number of data points, but in the right limit, this will give you the true distribution of parameters. (See the Hogg paper!)
